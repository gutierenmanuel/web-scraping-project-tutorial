### Countries of the World to dataframe

```python
# selenium 4
from selenium import webdriver
from selenium.webdriver.firefox.service import Service as FirefoxService
from webdriver_manager.firefox import GeckoDriverManager

driver = webdriver.Firefox(service=FirefoxService(GeckoDriverManager().install()))
```

```python
import time
from selenium.webdriver.common.by import By
import pandas as pd
```

```python

driver = webdriver.Firefox(service=FirefoxService(GeckoDriverManager().install()))
url = "https://www.scrapethissite.com/pages/simple/"
driver.get(url)

time.sleep(3)
```

```python
countries = {}
```

```python
country_divs = driver.find_elements(By.CLASS_NAME,'country')

for country in country_divs:
    
    name = country.find_element(By.CLASS_NAME,'country-name').text
    capital = country.find_element(By.CLASS_NAME,'country-capital').text
    population = country.find_element(By.CLASS_NAME,'country-population').text
    area = country.find_element(By.CLASS_NAME,'country-area').text

    countries[name] = {
        'capitals': capital,
        'populations': population,
        'areas': area
    }
```

```python
# Configura Pandas para mostrar DataFrames completos sin truncar
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)
```

```python
# Restaura el valor de pandas para mostrar dataframes truncados
pd.reset_option('all')
```

```python
df = pd.DataFrame.from_dict(countries, orient='index')
df
```
